```python
fb=pd.read_csv('D:\\ALL DATA SCIENCE\\Learn Digital\\DataSets\\fb_sentiment.csv')
```

```python
fb
```

```python
fb.drop('Unnamed: 0', axis=1, inplace=True)
fb.head()
```

```python
fb=[x.strip() for x in fb.FBPost]
fb=[x for x in fb if x]
fb[0:5]
```

```python
# Joining the list into one string/text
text = ' '.join(fb)
text
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
words = nltk.word_tokenize(text)
words
```

```python
no_punct= [word for word in words if word.isalnum()]
no_punct
```

```python
lower_words=[x.lower() for x in no_punct]
lower_words
```

```python
len(lower_words)
```

```python
#Remove stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
my_stop_words = stopwords.words('english')
my_stop_words
```

```python
no_stop_tokens = [word for word in lower_words if not word in my_stop_words]
```

```python
no_stop_tokens
```

```python
len(no_stop_tokens)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
for w in no_stop_tokens:
    print(w, " : ", ps.stem(w))
```

```python
stemmed_words=[ps.stem(w) for w in no_stop_tokens]
stemmed_words
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
for w in stemmed_words:
    print(w, " : ", lemmatizer.lemmatize(w))
```
