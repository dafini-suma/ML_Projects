```python
import pandas as pd
import numpy as np
df_iris = pd.read_csv("C:\\Users\\User\\Downloads\\iris.csv")
df_iris
```

```python
df_slen=pd.DataFrame(df_iris['Sepal.Length'])
l1=list(df_slen['Sepal.Length'])
df_mean=sum(l1) / len(l1)
count=len(df_slen)
x=[0.0]*150
xsqr=[0.0]*150

for i in range(0,count):
    x[i]=l1[i]-df_mean
    xsqr[i]=x[i]*x[i]
df_slen["x"]=x
df_slen["x^2"]=xsqr
df_slen
```

```python
s1=df_slen["x^2"].sum()
s1
```

```python
#variance
var1=s1/(count-1)
var1
```

```python
#standard deviation
stdv=var1**0.5
stdv
```
