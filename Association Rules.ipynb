```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
titanic = pd.read_csv("F:\\Data Science\\Data Sets\\Titanic_Apriori.csv")
titanic
```

```python
df = pd.get_dummies(titanic)
df
```

```python
titanic.dtypes
```

```python
df.dtypes
```

```python
frequent_itemsets = apriori(df,min_support=0.4,use_colnames=True)
frequent_itemsets
```

```python
rules =association_rules(frequent_itemsets, metric = "lift",min_threshold=0.7)
rules
```

```python
rules.sort_values("lift",ascending=False).head(15)
```

```python
claimants = pd.read_csv("F:\\Data Science\\Data Sets\\claimants.csv")
claimants
```

```python
claimants.isnull().sum()
```

```python
claimants.shape
```

```python
claimants = claimants.dropna()
```

```python
claimants.shape
```

```python
claimants['CLMAGE1']=pd.cut(x = claimants['CLMAGE'], bins = [-1,25,50,100],right=True,labels = ['Student','Working','Aged'])
claimants
```

```python
claimants = pd.get_dummies(claimants, columns=['CLMAGE1'])
claimants
```

```python
claimants['LOSS1']=pd.cut(x = claimants['LOSS'], bins = [-1,50,70,174],right=True,labels = ['Low Claim','Moderate Claim','High Claim'])
claimants
```

```python
claimants.isnull().sum()
```

```python
claimants = pd.get_dummies(claimants, columns=['LOSS1'])
claimants
```

```python
claimants.dtypes
```

```python

claimants['ATTORNEY NOT APPOINTED']=np.where(claimants['ATTORNEY']==0,'0','1')
claimants['ATTORNEY APPOINTED']=np.where(claimants['ATTORNEY']==0,'1','0')
claimants
```

```python
claimants['Male']=np.where(claimants['CLMSEX']==0,'1','0')
claimants['Female']=np.where(claimants['CLMSEX']==0,'0','1')
claimants
```

```python
claimants['Wearing Seatbelt']=np.where(claimants['SEATBELT']==0,'1','0')
claimants['Not Wearing Seatbelt']=np.where(claimants['SEATBELT']==0,'0','1')
claimants
```

```python
claimants['Insured']=np.where(claimants['CLMINSUR']==0,'0','1')
claimants['Not Insured']=np.where(claimants['CLMINSUR']==0,'1','0')
claimants
```

```python
claimants.dtypes
```

```python
claimants_new = claimants
claimants_new
```

```python
claimants_new.dtypes
```

```python
claimants_new = claimants_new.drop('CASENUM',axis = 1)
claimants_new = claimants_new.drop('CLMSEX',axis = 1)
claimants_new = claimants_new.drop('CLMINSUR',axis = 1)
claimants_new = claimants_new.drop('SEATBELT',axis = 1)
claimants_new = claimants_new.drop('CLMAGE',axis = 1)
claimants_new = claimants_new.drop('ATTORNEY',axis = 1)
claimants_new = claimants_new.drop('LOSS',axis = 1)
claimants_new
```

```python
claimants_new.dtypes
```

```python
claimants_new['ATTORNEY NOT APPOINTED'] = claimants_new['ATTORNEY NOT APPOINTED'].astype(int)
claimants_new['ATTORNEY APPOINTED'] = claimants_new['ATTORNEY APPOINTED'].astype(int)
claimants_new['Male'] = claimants_new['Male'].astype(int)
claimants_new['Female'] = claimants_new['Female'].astype(int)
claimants_new['Wearing Seatbelt'] = claimants_new['Wearing Seatbelt'].astype(int)
claimants_new['Not Wearing Seatbelt'] = claimants_new['Not Wearing Seatbelt'].astype(int)
claimants_new['Insured'] = claimants_new['Insured'].astype(int)
claimants_new['Not Insured'] = claimants_new['Not Insured'].astype(int)
#claimants['SEATBELT'] = claimants['SEATBELT'].astype(int)
```
