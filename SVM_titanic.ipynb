```python
data['age'] = data['age'].fillna(data.groupby('sibsp')['age'].transform('mean'))
data
```

```python
data = data.drop('name',axis = 1)
data
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder()
data['sex'] = label_encoder.fit_transform(data['sex'])
data['embarked'] = label_encoder.fit_transform(data['embarked'])
data
```

```python
#X=data.iloc[:,2:]
X = pd.concat([data.iloc[:,0],data.iloc[:,2:8]],axis = 1)
Y=data['survived']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
clf = SVC()
#param_grid = [{'kernel':['poly'],'gamma':[1, 0.1, 0.01, 0.001, 0.0001],'C':[0.1, 1, 10, 100, 1000]}]
param_grid = [{'kernel':['rbf'],'gamma':[1, 0.1, 0.01, 0.001, 0.0001],'C':[0.1, 1, 10, 100, 1000]}]
gsv = GridSearchCV(clf,param_grid,refit = True, verbose = 3)
gsv.fit(X_train,Y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 100, gamma = 0.001)
clf.fit(X_train , Y_train)
y_pred_train = clf.predict(X_train)
acc = accuracy_score(Y_train, y_pred_train) * 100
print("Accuracy =", acc)
confusion_matrix(Y_train, y_pred_train)
```

```python
y_pred_test = clf.predict(X_test)
acc = accuracy_score(Y_test, y_pred_test) * 100
print("Accuracy =", acc)
confusion_matrix(Y_test, y_pred_test)
```
