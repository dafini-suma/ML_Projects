```python
# Create your first MLP in Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
# Fix random seed for reproducibility
numpy.random.seed(7)
# load pima indians dataset
dataset = numpy.loadtxt("D:\\ALL DATA SCIENCE\\Learn Digital\\DataSets\\pima-indians-diabetes.csv", 
                        delimiter=",")
dataset
```

```python
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
X[0:5]
```

```python
Y[0:5]
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8,  activation='relu')) #1st layer
model.add(Dense(8,  activation='relu')) #2nd layer
model.add(Dense(1, activation='sigmoid')) #3rd layer or top layer

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
history=model.fit(X, Y, validation_split=0.33, epochs=250, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print(scores)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
